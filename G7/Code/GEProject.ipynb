##เพิ่มคอลัมน์เพศ

```python
def infer_gender(prefix):
    if prefix == 'นาย':
        return 'ชาย'
    elif prefix in ['นาง', 'นางสาว']:
        return 'หญิง'
    else:
        return 'ไม่ระบุ'  # เผื่อมีคำนำหน้าอื่น

df['เพศ'] = df['คำนำหน้า'].apply(infer_gender)
```

##ใช้ OridnalEncoder และ LabelEncoder แปลง

```python
# ประเภทบุคคล (ไม่มีลำดับ)
label_encoder_person_type = LabelEncoder()
df['ประเภทบุคคล_encoded'] = label_encoder_person_type.fit_transform(df['ประเภทบุคคล'])

# ช่วงเวลาที่ออกกำลังกาย (ไม่มีลำดับ)
label_encoder_time = LabelEncoder()
df['ช่วงเวลา_encoded'] = label_encoder_time.fit_transform(df['ช่วงเวลาที่ออกกำลังกายเป็นประจำ'])
```

```python
# คำนำหน้า (มีลำดับ)
prefix_order = [['นาย', 'นาง', 'นางสาว']]
ordinal_encoder_prefix = OrdinalEncoder(categories=prefix_order)
df['คำนำหน้า_encoded'] = ordinal_encoder_prefix.fit_transform(df[['คำนำหน้า']]).astype(int)

# เพศ (มีลำดับ)
gender_order = [['ชาย', 'หญิง']]
ordinal_encoder_gender = OrdinalEncoder(categories=gender_order)
df['เพศ_encoded'] = ordinal_encoder_gender.fit_transform(df[['เพศ']]).astype(int)

# อายุ (มีลำดับ)
age_order = [['10-14 ปี', '15-19 ปี', '20-24 ปี', '25-29 ปี', '30-34 ปี', '35 ปีขึ้นไป']]
ordinal_encoder_age = OrdinalEncoder(categories=age_order)
df['อายุ_encoded'] = ordinal_encoder_age.fit_transform(df[['อายุ']]).astype(int)

# ระยะเวลาออกกำลังกายเฉลี่ยต่อครั้ง (มีลำดับ)
duration_order = [['น้อยกว่า 1 ชม.', '1-2 ชม.', '3-5 ชม.', '5-10 ชม.']]
ordinal_encoder_duration = OrdinalEncoder(categories=duration_order)
df['ระยะเวลา_encoded'] = ordinal_encoder_duration.fit_transform(df[['ระยะเวลาออกกำลังกายเฉลี่ยต่อครั้ง']]).astype(int)

# รายชื่อคอลัมน์ของปัจจัย
factor_columns = [
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อสุขภาพร่างกาย]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อสุขภาพจิต/ลดความเครียด]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อควบคุมน้ำหนัก/รูปร่าง]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อเพิ่มสมรรถภาพ/สมาธิ]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อฟื้นฟูร่างกาย/กายภาพบำบัด]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เพื่อนชวน/ทำร่วมกับเพื่อน]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ได้พบปะ/เข้าสังคม]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ครอบครัว/คนรอบตัวสนับสนุน]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [สถานที่เอื้อต่อการออกกำลังกาย]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [เข้าร่วมแข่งขัน/ชมรม/ทีมกีฬา]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ตามกระแสสังคม/สื่อโซเชียล]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [สนุก/ทำเพราะชอบ]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ความสะดวกเรื่องเวลา]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ใช้เวลาให้เกิดประโยชน์]',
    'ปัจจัยสำคัญที่ทำให้คุณเลือกออกกำลังกาย [ความท้าทายส่วนตัว/ตั้งเป้าหมาย]'
]

# ระดับของปัจจัยจากน้อยที่สุด -> มากที่สุด (ใช้เหมือนกันทุกคอลัมน์)
factor_levels = [['น้อยที่สุด', 'น้อย', 'ปานกลาง', 'มาก', 'มากที่สุด']] * len(factor_columns)

# สร้างและใช้ OrdinalEncoder สำหรับปัจจัย
ordinal_encoder_factors = OrdinalEncoder(categories=factor_levels)
df[factor_columns] = ordinal_encoder_factors.fit_transform(df[factor_columns]).astype(int)

# Mapping ค่าระดับของปัจจัย
factor_mapping = dict(zip(factor_levels[0], range(len(factor_levels[0]))))
```

```python
allowed_sports_by_location = {
    "สนามเขียว (บริเวรข้างตลาดมอ)": ["ฟุตซอล", "วอลเลย์บอล"],
    "สวนเกษตร": ["แบดมินตัน", "วอลเลย์บอล", "เดิน/วิ่ง", "ปั่นจักรยาน"],
    "สระพลาสติก": ["กระโดดเชือก", "เดิน/วิ่ง"],
    "บึงศรีฐาน": ["กระโดดเชือก", "มวย (ต่อยมวย)", "เดิน/วิ่ง", "ปั่นจักรยาน"],
    "ยิมใหม่ (sport complex)": [
        "กระโดดเชือก", "เทควันโด", "เทเบิลเทนนิส (ปิงปอง)", "บาสเกตบอล",
        "แบดมินตัน", "ฟุตบอล", "ฟุตซอล", "มวย (ต่อยมวย)", "ยิมนาสติก",
        "ยูโด", "วอลเลย์บอล", "เซปักตะกร้อ", "โยคะ", "แอโรบิก"
    ],
    "ยิมเก่า (อาคารพลศึกษา)": [
        "กระโดดเชือก", "คาลิสเทนิกส์/บอดี้เวท (ดันพื้น/ดึงข้อ/สควอต)",
        "มวย (ต่อยมวย)", "เวทเทรนนิง (ยกน้ำหนัก/เครื่องออกกำลังกาย)", "เดิน/วิ่ง"
    ],
    "บริเวรสนาม 50 ปี (รวมถึง สระว่ายน้ำมหาลัย, สนามบาสสี่สนาม, สนามเทนนิส, สนามเปตอง)": [
        "กระโดดเชือก", "คาลิสเทนิกส์/บอดี้เวท (ดันพื้น/ดึงข้อ/สควอต)", "เทนนิส",
        "บาสเกตบอล", "แบดมินตัน", "ฟุตบอล", "ว่ายน้ำ", "เซปักตะกร้อ", "เปตอง", "เดิน/วิ่ง"
    ]
}
```

```python
def prepare_sports_column(df):
    """แยกกีฬาออกจาก string เป็น list และ explode เป็นแถว"""
    df['กิจกรรม/กีฬาที่ออกกำลังกาย'] = df['กิจกรรม/กีฬาที่ออกกำลังกาย'].str.split(',')
    df = df.explode('กิจกรรม/กีฬาที่ออกกำลังกาย')
    df['กิจกรรม/กีฬาที่ออกกำลังกาย'] = df['กิจกรรม/กีฬาที่ออกกำลังกาย'].str.strip()
    return df

def is_valid_sport(row, allowed_sports_by_location):
    location = row['สถานที่ที่ออกกำลังกายเป็นประจำ']
    sport = row['กิจกรรม/กีฬาที่ออกกำลังกาย']
    allowed_sports = allowed_sports_by_location.get(location, [])
    return sport in allowed_sports

def filter_valid_sports_by_location(df, allowed_sports_by_location):
    """รับ DataFrame แล้วกรองเฉพาะกีฬาที่ตรงกับสถานที่"""
    df_prepared = prepare_sports_column(df)
    df_valid = df_prepared[df_prepared.apply(lambda row: is_valid_sport(row, allowed_sports_by_location), axis=1)].copy()
    df_invalid = df_prepared[~df_prepared.apply(lambda row: is_valid_sport(row, allowed_sports_by_location), axis=1)].copy()
    return df_valid.reset_index(drop=True), df_invalid.reset_index(drop=True)
```
